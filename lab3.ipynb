```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal as sg
```

```python
# write a code to retrive the continous signal from sampled signal using 
# 1) Zero Order Hold (Z0H)
# 2) first order hold (f0H)
# x(t) = sin(w1t) + sin(w2t)

# w1 = 2*pi*f1 f1 = 2
# w2 = 2*pi*f2 f2 = 5
```

```python
#genrating signal 
f1 = 2
w1  = 2*(np.pi)*f1
f2 = 5
w2 = 2*(np.pi)*f2

Fs = 100    #sample rate- no of samples per second
T = 1/Fs    #sampling period   %sample per second
t = 1       #seconds of sampling
N = Fs*t    #total points in signal

t = np.arange(N)*T
x = np.sin(w1*t) + np.sin(w2*t)
plt.plot(x,label ='signal', color='green',  linestyle='dashed',
     linewidth=2, markersize=12)
plt.grid()
plt.title('input siganl',color='b')

```

```python
from scipy import signal as sg
#Zero order Hold

# step-1 genrating rectangular pulse
f1 = 2
w1  = 2*(np.pi)*f1
t = np.arange(0,1,0.01)

rect = []

for i in t:
    if(i < 0.1):
        rect.append(1)
    else:
        rect.append(0)

plt.plot(t,rect)
plt.grid()

xzoh = np.convolve(x,rect)

plt.title('rectangular pulse',color='b')
```

```python
Ts = 0.001
t = np.arange(0,1,Ts)

tringular = []
for i in t:
    if(i < Ts ):
        tringular.append(i)
    elif(i<2*Ts):
        tringular.append((1-i))
    else:
        tringular.append(0)



plt.plot(t, tringular)
plt.grid()
```

```python
fzoh = np.convolve(x,tringular)
plt.plot(fzoh[0:100])
plt.title('first order hold interpolation',color='b')
```
