```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 20
w = np.arange(0,2*np.pi,0.001)
u =  (1- np.exp(-1j*a*w))
d = (1- np.exp(-1j*w))
x = u/d

x_real = np.real(x)
x_img = np.imag(x)
x_abs = np.abs(x)

plt.plot(w,x_abs)
plt.grid()
plt.title('x_dtft siganl',color='b')
```
